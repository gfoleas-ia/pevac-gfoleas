## 1. Descripción del modelo

BLIP es un modelo de visión-lenguaje. En este taller se usa para una tarea de **image captioning**, es decir, generar texto a partir de una imagen. La aplicación es educativa y forma parte del enfoque SOPP+IA Sofía para entrenamiento en inteligencia artificial aplicada a ortopedia pediátrica.

```python
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
from google.colab import files
import matplotlib.pyplot as plt
import torch
```

```python
modelo = "Salesforce/blip-image-captioning-base"

processor = BlipProcessor.from_pretrained(modelo)
model = BlipForConditionalGeneration.from_pretrained(modelo)

print("Modelo cargado correctamente:", modelo)
```

## 2. Cargar una imagen no identificable

Suba una imagen de pies infantiles sin rostro, nombre, documento u otros datos identificables.

```python
uploaded = files.upload()

nombre_imagen = list(uploaded.keys())[0]
imagen = Image.open(nombre_imagen).convert("RGB")

plt.imshow(imagen)
plt.axis("off")
plt.title("Imagen cargada")
plt.show()
```

```python
inputs = processor(imagen, return_tensors="pt")

with torch.no_grad():
    output = model.generate(**inputs, max_new_tokens=30)

descripcion = processor.decode(output[0], skip_special_tokens=True)

print("Descripción automática generada por el modelo:")
print(descripcion)
```

```python
print("Interpretación educativa:")
print("El modelo genera una descripción general de la imagen.")
print("En el contexto de pie equino varo, puede servir para documentar visualmente una imagen clínica no identificable.")
print("No está entrenado para diagnóstico ortopédico, no calcula escala de Pirani y no reemplaza la valoración médica.")
```

## 3. Prueba con varias imágenes

Esta celda permite subir dos o más imágenes y generar una descripción para cada una.

```python
uploaded = files.upload()

for nombre_imagen in uploaded.keys():
    imagen = Image.open(nombre_imagen).convert("RGB")
    inputs = processor(imagen, return_tensors="pt")
    
    with torch.no_grad():
        output = model.generate(**inputs, max_new_tokens=30)
    
    descripcion = processor.decode(output[0], skip_special_tokens=True)
    
    plt.imshow(imagen)
    plt.axis("off")
    plt.title(nombre_imagen)
    plt.show()
    
    print("Resultado:")
    print(descripcion)
    print("-" * 50)
```

## 4. Aplicación adicional con Gradio

Esta parte permite demostrar el modelo en una interfaz sencilla dentro de Google Colab.

```python
import gradio as gr

def describir_imagen_gradio(imagen):
    imagen = Image.fromarray(imagen).convert("RGB")
    inputs = processor(imagen, return_tensors="pt")
    with torch.no_grad():
        output = model.generate(**inputs, max_new_tokens=30)
    descripcion = processor.decode(output[0], skip_special_tokens=True)
    return descripcion

app = gr.Interface(
    fn=describir_imagen_gradio,
    inputs=gr.Image(type="numpy", label="Subir imagen no identificable del pie infantil"),
    outputs=gr.Textbox(label="Descripción automática generada"),
    title="SOPP+IA Sofía - Pie Equino Varo",
    description="Aplicación educativa de entrenamiento. No diagnostica, no calcula Pirani y no reemplaza la valoración médica."
)

app.launch(share=True)
```
